### Data Cleaning


```python
import re

def clean_medical_text(text):

    # remove video instructions
    text = re.sub(r"watch the video.*?keyboard\.", "", text, flags=re.I)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text)

    # remove weird characters
    text = re.sub(r"[^\w\s.,-]", "", text)

    return text.strip()
```

```python
df["clean_answer"] = df["answer"].apply(clean_medical_text)
```

```python
df[["answer","clean_answer"]].iloc[30:40]
```

```python
def clean_question(q):

    q = q.lower()
    q = re.sub(r"\(.*?\)", "", q)
    q = re.sub(r"\s+", " ", q)

    return q.strip()

df["clean_question"] = df["question"].apply(clean_question)
```

```python
df_clean = df[[
    "clean_question",
    "clean_answer",
    "focus_area",
    "source"
]]

df_clean.to_csv("data/medquad_cleaned.csv", index=False)
```

### Chunk Long Documents

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
#Creating the splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=150
)
```

```python
# Splitting the document
documents = []

for i, row in df_clean.iterrows():

    chunks = text_splitter.split_text(row["document"])

    for chunk in chunks:
        documents.append({
            "text": chunk,
            "focus_area": row["focus_area"],
            "source": row["source"]
        })
```

```python
chunk_df = pd.DataFrame(documents)
chunk_df.head()
```

### Generating Embeddings

```python
#Importing the embedding model
from sentence_transformers import SentenceTransformer

# We will use a lightweight but strong model:all-MiniLM-L6-v2
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
#Extract the text column.
texts = chunk_df["text"].tolist()
```

```python
texts[0]
```

```python
# Generate Embeddings to convert each chunk into a vector representation
embeddings = embedding_model.encode(
    texts,
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)
```

```python
embeddings.shape
```

```python
import faiss
import numpy as np

#Build the vector Index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
index.ntotal
```

```python
#create the models folder
os.makedirs("models", exist_ok=True)

# Save the Vector Database
faiss.write_index(index, "models/medquad_faiss.index")
chunk_df.to_csv("models/medquad_metadata.csv", index=False)
```

```python
#Test Semantic Search
#create a function
def search(query, top_k=5):

    query_embedding = embedding_model.encode([query])

    distances, indices = index.search(query_embedding, top_k)

    results = chunk_df.iloc[indices[0]]

    return results
```

```python
# Test
results = search("What causes glaucoma?")
results["text"]
```

The system takes the user's query, converts it into an embedding, searches the FAISS vector database, and retrieves the most relevant medical knowledge chunks. However, at this stage the system only returns the retrieved context and does not yet generate a final answer using an LLM.

```python
def ask_medical_question(query):

    # Retrieve relevant chunks
    results = search(query, top_k=2)

    # Join retrieved chunks
    context_chunks = results["text"].tolist()
    context = "\n".join(context_chunks)

    # Limit context size to avoid token overflow
    context = context[:1500]

    # Extract unique sources
    sources = ", ".join(results["source"].unique())

    prompt = f"""
You are a medical information assistant.

Use ONLY the medical context below to answer the question clearly and concisely.

If the answer cannot be found in the context, say:
"The available medical sources do not provide this information."

Do not provide medical diagnosis or treatment advice.

Context:
{context}

Question:
{query}

Answer:
"""

    response = llm(prompt)

    # Clean output
    answer = response[0]["generated_text"].replace(prompt, "").strip()

    return {
        "answer": answer,
        "sources": sources
    }
```

```python
import pkg_resources
from symspellpy import SymSpell
from sentence_transformers import CrossEncoder
```

```python
#initialize spell correction(Symspell)
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)

dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)

sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
```

```python
#Load cross-Encoder Reranker
reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")
```

```python
#query normalization
def normalize_query(query):

    suggestions = sym_spell.lookup_compound(query, max_edit_distance=2)

    if suggestions:
        return suggestions[0].term
    else:
        return query
```

```python
#query cleaning
def clean_query(query):

    query = query.lower()
    query = re.sub(r"[^a-zA-Z0-9\s]", "", query)
    query = query.strip()

    return query
```

```python
#FAISS Retrieval
def search(query, top_k=10):

    query_embedding = embedding_model.encode([query])

    distances, indices = index.search(query_embedding, top_k)

    results = chunk_df.iloc[indices[0]]

    results = results.drop_duplicates(subset="text")

    return results, distances[0]
```

```python
#Cross-Encoder Reranking
def rerank_results(query, results, top_k=3):

    pairs = [(query, text) for text in results["text"]]

    scores = reranker.predict(pairs)

    results = results.copy()
    results["score"] = scores

    results = results.sort_values("score", ascending=False)

    return results.head(top_k)
```

```python
#Context filtering
def filter_context(results, query):

    keywords = query.lower().split()

    filtered = results[
        results["text"].str.lower().apply(
            lambda x: any(word in x for word in keywords)
        )
    ]

    if len(filtered) == 0:
        filtered = results

    return filtered
```

```python
#answer formatting
def format_answer(text):

    sentences = re.split(r'(?<=[.!?]) +', text)

    sentences = [s.capitalize() for s in sentences]

    return " ".join(sentences)
```

```python
#RAG Function
def ask_medical_question(query):

    # spelling correction
    query = normalize_query(query)

    # clean query
    query = clean_query(query)

    # retrieve candidate chunks
    results, distances = search(query, top_k=10)

    # hallucination guard
    if min(distances) > 1.2:
        return "The available medical sources do not provide this information."

    # rerank retrieved chunks
    results = rerank_results(query, results, top_k=3)

    # filter context
    results = filter_context(results, query)

    # build context
    context = "\n".join(results["text"].tolist())
    context = context[:1200]

    sources = ", ".join(results["source"].unique())

    prompt = f"""
You are a medical information assistant.

Answer the question using ONLY the medical context below.

If the context does not contain the answer, say:
"The available medical sources do not provide this information."

Context:
{context}

Question:
{query}

Answer in 2вЂ“3 clear sentences:
"""

    response = llm(prompt)

    answer = response[0]["generated_text"].strip()

    answer = format_answer(answer)

    return f"{answer}\n\nSource: {sources}"
```
